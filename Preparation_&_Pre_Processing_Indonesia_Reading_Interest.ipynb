```python
# 1. Import the necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler

# Ensure plots display inline in Colab
%matplotlib inline
```

```python
# 2. Read the dataset
df = pd.read_csv("/content/TGM 2020-2023_eng.csv", sep=";")

# Print detected column names
print("=== Detected Columns ===")
print(df.columns.tolist(), "\n")
```

```python
# 3. Display the first few rows to verify the structure
print("=== First 5 Rows ===")
print(df.head(), "\n")
```

```python
# 4. Basic exploration: info, describe, missing values
print("=== DataFrame Info ===")
print(df.info(), "\n")
print("=== Descriptive Statistics ===")
print(df.describe(include='all'), "\n")
print("=== Missing Values Per Column ===")
print(df.isnull().sum(), "\n")
```

```python
# 5. Remove duplicate rows (if any)
duplicates = df.duplicated().sum()
print("Number of duplicate rows:", duplicates)
if duplicates > 0:
    df = df.drop_duplicates()
    print("After removing duplicates, shape:", df.shape, "\n")
else:
    print("No duplicate rows found.\n")
```

```python
# 6. Handle missing values and convert columns that contain commas to proper numeric
# Define the reading interest column
reading_col = "Tingkat Kegemaran Membaca (Reading Interest)"

# List of columns with numeric values stored as strings (with commas)
numeric_cols_with_commas = [
    "Reading Frequency per week",
    "Number of Readings per Quarter",
    "Daily Reading Duration (in minutes)",
    "Internet Access Frequency per Week",
    "Daily Internet Duration (in minutes)",
    reading_col
]

for col in numeric_cols_with_commas:
    if col in df.columns:
        # Replace commas with dots
        df[col] = df[col].astype(str).str.replace(',', '.')
        # Convert to numeric
        df[col] = pd.to_numeric(df[col], errors='coerce')
        # Fill missing with median
        median_val = df[col].median()
        df[col] = df[col].fillna(median_val)
        print(f"Missing values in '{col}' filled with median: {median_val}")

# Convert "Year" to numeric if it exists
if 'Year' in df.columns:
    df['Year'] = pd.to_numeric(df['Year'], errors='coerce')
    year_median = df['Year'].median()
    df['Year'] = df['Year'].fillna(year_median)
    print(f"Missing values in 'Year' filled with median: {year_median}")

# Fill missing values in categorical columns with mode
categorical_cols = ["Provinsi", "Category"]
for cat_col in categorical_cols:
    if cat_col in df.columns:
        mode_val = df[cat_col].mode()[0]
        df[cat_col] = df[cat_col].fillna(mode_val)
        print(f"Missing values in '{cat_col}' filled with mode: {mode_val}")
print()
```

```python
# 7. Visualize the reading interest BEFORE normalization
plt.figure(figsize=(8, 6))
sns.histplot(df[reading_col], bins=20, kde=True)
plt.title(f"Distribution of '{reading_col}' (Before Outlier Removal & Normalization)")
plt.xlabel(reading_col)
plt.ylabel("Frequency")
plt.show()
```

```python
# 8. Detect and remove outliers using the IQR method
def remove_outliers_iqr(data, cols):
    data_clean = data.copy()
    for c in cols:
        Q1 = data_clean[c].quantile(0.25)
        Q3 = data_clean[c].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        outlier_indices = data_clean[(data_clean[c] < lower_bound) | (data_clean[c] > upper_bound)].index
        print(f"Column '{c}': {len(outlier_indices)} outliers detected.")
        data_clean = data_clean.drop(index=outlier_indices)
    return data_clean

# Identify numeric columns for outlier detection
numeric_cols = df.select_dtypes(include=[np.number]).columns.tolist()
print("Numeric columns:", numeric_cols, "\n")

# Remove outliers from numeric columns
df_no_outliers = remove_outliers_iqr(df, numeric_cols)
print("\nShape after outlier removal:", df_no_outliers.shape)

# Visualize the reading interest AFTER outlier removal
plt.figure(figsize=(8, 6))
sns.histplot(df_no_outliers[reading_col], bins=20, kde=True, color='orange')
plt.title(f"Distribution of '{reading_col}' (After Outlier Removal)")
plt.xlabel(reading_col)
plt.ylabel("Frequency")
plt.show()
```

```python
# 9. Normalize the data AFTER outlier removal using MinMaxScaler
df_no_outliers_normalized = df_no_outliers.copy()
scaler = MinMaxScaler()
# Re-identify numeric columns in the outlier-removed dataset
numeric_cols_no_outliers = df_no_outliers_normalized.select_dtypes(include=[np.number]).columns.tolist()
df_no_outliers_normalized[numeric_cols_no_outliers] = scaler.fit_transform(df_no_outliers_normalized[numeric_cols_no_outliers])

# Visualize the reading interest AFTER normalization
plt.figure(figsize=(8, 6))
sns.histplot(df_no_outliers_normalized[reading_col], bins=20, kde=True, color='purple')
plt.title(f"Distribution of '{reading_col}' (After Normalization on Outlier-Free Data)")
plt.xlabel(f"{reading_col} (Normalized)")
plt.ylabel("Frequency")
plt.show()

print("\nSample of normalized data (after outlier removal):")
print(df_no_outliers_normalized.head())
```

```python
# 10. Save the preprocessed datasets
df_no_outliers_normalized.to_csv("TGM_2020-2023_normalized.csv", index=False)
df_no_outliers.to_csv("TGM_2020-2023_cleaned.csv", index=False)
print("\nNormalized dataset (outlier-free) saved as 'TGM_2020-2023_normalized.csv'.")
print("Cleaned (no outliers) dataset saved as 'TGM_2020-2023_cleaned.csv'.")
```

```python
# === Airline dataset visualizations ===

import io
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams.update({
    "figure.figsize": (9, 6),
    "axes.grid": True,
    "grid.linestyle": ":",
    "grid.alpha": 0.4
})

# --- Set your path here ---
# Example: "/content/drive/MyDrive/Colab Notebooks/Airline Dataset Updated - v2.csv"
file_path = "/content/Airline Dataset Updated - v2.csv"

# --- Load CSV ---
df = pd.read_csv(file_path)

# Standardize column names (strip spaces)
df.columns = [c.strip() for c in df.columns]

# Parse dates if present
date_cols = []
for col in ["Departure Date", "Arrival Date", "DepartureDate", "ArrivalDate"]:
    if col in df.columns:
        date_cols.append(col)

for c in date_cols:
    df[c] = pd.to_datetime(df[c], errors="coerce", infer_datetime_format=True)

# Preview
display(df.head())
print(df.info())

# ====== Helper plotting function ======
def show_bar(series, title, xlabel="", ylabel="Count", top=None, rot=45):
    data = series.value_counts().sort_values(ascending=False)
    if top:
        data = data.head(top)
    ax = data.plot(kind="bar", edgecolor="black")
    ax.set_title(title)
    ax.set_xlabel(xlabel)
    ax.set_ylabel(ylabel)
    plt.xticks(rotation=rot, ha="right")
    for i, v in enumerate(data.values):
        ax.text(i, v, f"{v:,}", va="bottom", ha="center", fontsize=9)
    plt.tight_layout()
    plt.show()

# =======================
# 1) Gender distribution
# =======================
if "Gender" in df.columns:
    gender_counts = df["Gender"].value_counts()
    plt.figure(figsize=(6,6))
    plt.pie(gender_counts.values, labels=gender_counts.index, autopct="%.1f%%", startangle=90)
    plt.title("Passenger Gender Distribution")
    plt.tight_layout()
    plt.show()

# =======================
# 2) Age distribution
# =======================
if "Age" in df.columns:
    ages = df["Age"].dropna()
    plt.hist(ages, bins=30, edgecolor="black")
    plt.title("Passenger Age Distribution")
    plt.xlabel("Age")
    plt.ylabel("Number of Passengers")
    plt.tight_layout()
    plt.show()

# ==========================================
# 3) Top 10 passenger nationalities (bar)
# ==========================================
if "Nationality" in df.columns:
    show_bar(df["Nationality"], "Top 10 Passenger Nationalities", "Nationality", top=10)

# ===================================================
# 4) Passengers by continent (from Country/Continent)
# ===================================================
continent_col = None
for c in ["Airport Continent", "Continents", "Continent"]:
    if c in df.columns:
        continent_col = c
        break

if continent_col:
    show_bar(df[continent_col], f"Passengers by Continent ({continent_col})", "Continent", rot=0)

# =============================================
# 5) Flight status distribution (On time, etc.)
# =============================================
if "Flight Status" in df.columns:
    show_bar(df["Flight Status"], "Flight Status Distribution", "Status", rot=0)

# ======================================================
# 6) Monthly passenger volume (by Departure Date month)
# ======================================================
if "Departure Date" in df.columns:
    dep = df.dropna(subset=["Departure Date"]).copy()
    dep["Month"] = dep["Departure Date"].dt.to_period("M").dt.to_timestamp()
    monthly = dep["Month"].value_counts().sort_index()
    ax = monthly.plot(kind="line", marker="o")
    ax.set_title("Passengers per Month (by Departure Date)")
    ax.set_xlabel("Month")
    ax.set_ylabel("Passengers")
    plt.tight_layout()
    plt.show()

# ==================================================================
# 7) On-time rate by Country (Top 12 countries by passenger volume)
# ==================================================================
if "Country Name" in df.columns and "Flight Status" in df.columns:
    tmp = df.copy()
    tmp["is_on_time"] = (tmp["Flight Status"].str.strip().str.lower() == "on time")
    by_country = tmp.groupby("Country Name")["is_on_time"].mean()
    top_countries = df["Country Name"].value_counts().head(12).index
    ontime_top = by_country.loc[top_countries].sort_values(ascending=False)

    ax = ontime_top.plot(kind="bar", edgecolor="black")
    ax.set_ylim(0, 1)
    ax.set_title("On-Time Rate by Country (Top 12 by Volume)")
    ax.set_xlabel("Country")
    ax.set_ylabel("On-Time Rate")
    plt.xticks(rotation=45, ha="right")
    for i, v in enumerate(ontime_top.values):
        ax.text(i, v, f"{v*100:.1f}%", va="bottom", ha="center", fontsize=9)
    plt.tight_layout()
    plt.show()

# ==========================================
# 8) Top 12 airports by passenger departures
# ==========================================
if "Airport Name" in df.columns:
    show_bar(df["Airport Name"], "Top 12 Airports by Passengers", "Airport", top=12)

# ======================================================
# 9) Age vs Flight Status (boxplot for 3+ categories)
# ======================================================
if "Age" in df.columns and "Flight Status" in df.columns:
    # Keep only categories with decent counts
    vc = df["Flight Status"].value_counts()
    keep = vc[vc >= 100].index
    filtered = df[df["Flight Status"].isin(keep) & df["Age"].notna()]

    if len(keep) >= 2:
        order = keep.tolist()
        data = [filtered.loc[filtered["Flight Status"] == k, "Age"].values for k in order]
        plt.boxplot(data, labels=order, showmeans=True)
        plt.title("Age vs Flight Status")
        plt.xlabel("Flight Status")
        plt.ylabel("Age")
        plt.tight_layout()
        plt.show()
```
